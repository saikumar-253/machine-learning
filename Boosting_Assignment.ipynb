**Question 1: What is Boosting in Machine Learning? Explain how it improves weak
learners?**

ans:**Boosting** in Machine Learning is an **ensemble** technique that sequentially combines many simple "**weak**" models (like shallow decision trees) to create one highly accurate "**strong**" model, primarily by training each new model to correct the errors (misclassifications) of the previous ones.

 **how it improves weak
learners**

- **weak** **learners** are improves by combines  many weak leaners  into the stong learners by improving the previous model.

**Question 2: What is the difference between AdaBoost and Gradient Boosting in terms
of how models are trained?**

ans:

- **Adaboost**



 1.**Weight Update Strategy:** Increase weights of misclassified sample so that the next learner focuses more on them.

 2.**Base learners :** AdaBoost uses simple decision trees with one split known as the decision stumps of weak learners.


  - **Gradient**

  1.**Weight Update Strategy:** Updates predictions by minimizing a loss function using the negative gradient.

  **2.Base learners:**Gradient Boosting can use a wide range of base learners such as decision trees and linear models.



**Question 3: How does regularization help in XGBoost?**

ans:In XGBoost, regularization plays a key role in controlling model complexity and preventing overfitting, which is especially important because boosting builds many trees sequentially that can easily overfit the training data.

**Question 6: Write a Python program to:**

-  Train an AdaBoost Classifier on the Breast Cancer dataset

-  Print the model accuracy


```python
#loading the Breast Cancer dataset
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
X = data.data
```

```python
y = data.target
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.20,random_state=42)
```

```python
#  Initialize of adaboost classifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
classifier = AdaBoostClassifier(n_estimators=100,random_state=42)
```

```python
classifier.fit(X_train,y_train)
```

```python
#predication

y_pred = classifier.predict(X_test)
```

```python
# Calculate and print accuracy
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

**Question 9: Write a Python program to:**

-  Train a CatBoost Classifier
-  Plot the confusion matrix using seaborn

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
pip install catboost
```

```python
# Create a synthetic dataset with categorical + numeric features
data = pd.DataFrame({
    "age": [25, 32, 47, 51, 62, 23, 36, 44, 53, 60],
    "income": [50000, 60000, 80000, 72000, 91000, 45000, 62000, 70000, 85000, 93000],
    "gender": ["M", "F", "M", "F", "M", "F", "F", "M", "F", "M"],  # categorical
    "region": ["North", "South", "East", "West", "East", "North", "South", "West", "East", "South"],  # categorical
    "default": [0, 0, 1, 0, 1, 0, 0, 1, 1, 1]  # target (loan default: 1 = default, 0 = no default)
})
```

```python
# Separate features and target
X = data.drop("default", axis=1)
y = data["default"]
```

```python
# Identify categorical feature indices
cat_features = ["gender", "region"]
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from catboost import CatBoostClassifier
```

```python
# Initialize CatBoost Classifier
model = CatBoostClassifier(
    iterations=300,
    learning_rate=0.1,
    depth=6,
    cat_features=cat_features,  # directly handle categorical features
    verbose=0,
    random_seed=42
)
```

```python
# Train model
model.fit(X_train, y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
```

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Plot using seaborn
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["No Default", "Default"],
            yticklabels=["No Default", "Default"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("CatBoost Classifier - Confusion Matrix (Categorical Features)")
plt.show()
```
