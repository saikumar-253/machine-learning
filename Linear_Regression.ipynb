**Question 1: What is Simple Linear Regression?**


**ans**:Simple Linear Regression is method to understand the relation between **independent** variable and **dependent** variable.


**Equation** : **Y** = mx + c

**Question 2: What are the key assumptions of Simple Linear Regression?**

ans:

 1.**Linearity**: The relationship between X and Y is linear
 2.**Independence** **of** **Errors**: The residuals(errors) are independent of each other.

 3.**Homoscedasticity** :constant variance of residuals

 4.**Normality** **of** **Errors**: The residuals(error) are normally distributed.

**Question 3: What is heteroscedasticity, and why is it important to address in regression model?**

**ans**:

The **Heteroscendasticity** means constant variance of the residuals is not present .

**Question 4: What is Multiple Linear Regression?**

ans:

Multiple Linear Regression is two or more **independent** **variable** and one **dependent** **variable**.


**Equation** : Y = m1x1 +m2x +c

**Question 6: Implement a Python program to fit a Simple Linear Regression model tothe following sample data**



```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating the dataset
dataset = pd.DataFrame({"X":[1, 2, 3, 4, 5],"Y":[2.1, 4.3, 6.1, 7.9, 10.2]
}
)
```

```python
dataset
```

```python
#Step - 1 : divide the dataset into X and Y

x = dataset[['X']]
y = dataset[['Y']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#Step -2 train and test split

x_train,x_test,y_train,y_test  =  train_test_split(x,y,test_size= 0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#step - 3 train  the model
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
print("slope :",model.coef_ )
```

```python
print('intercept:',model.intercept_)
```

```python
#Plot the regression line over the data points

plt.scatter(x_train,y_train,label = "Data points")
plt.plot(x_train,model.predict(x_train),label = "Regression Line ",color = "red")
plt.xlabel("X")
plt.ylabel("Y")
plt.title('Simple Linear Regression ')
plt.legend()
```
