**Q2. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different
app categories.**

```python
plt.figure(figsize=(35,15))
sns.boxplot(x = "Category",y = "Rating", data  = df)
plt.title('Category wise Rating Boxplot')
plt.xticks(rotation = 45,ha = 'right')
plt.xlabel('Category')
plt.ylabel('Rating')
plt.show()
```

# observation

 - Art_and_Design category not present outliers and remain all are present.

**Q4. What is the relationship between the size of an app and its rating? Create a scatter plot to visualize
the relationship.**

```python
plt.figure(figsize=(10,6))
plt.scatter(x = "Size",y = "Rating",data = df)
plt.title("Size VS Rating")
plt.xlabel('Size')
plt.ylabel('Rating')
plt.show()
```
